```python
from google.colab import userdata
```

```python
import os
os.environ["PYTORCH_CUDA_ALLOC_CONF"] = "expandable_segments:True"
```

```python
from huggingface_hub import InferenceClient

client = InferenceClient(api_key=userdata.get('HF_TOKEN'))
```

```python
import torch
```

```python
import pandas as pd
df = pd.read_csv('SoruA-15chunk.csv')

template_value = df.loc[0, 'template']
split_values = template_value.split('\n\n\n\n')
new_df = pd.DataFrame(split_values, columns=['template'])
```

```python
for i in range(len(new_df["template"])):
  torch.cuda.empty_cache()  # Belleği temizle
  messages = [
    {"role": "user", "content": str(new_df["template"].iloc[i])},
  ]
  try:

    completion = client.chat.completions.create(
      model="google/gemma-2-9b-it",
    messages=messages,
    temperature=0.6,
    max_tokens=256,
    top_p=0.9
    )

    print(completion.choices[0].message.content)
  except:
    try:
      completion = client.chat.completions.create(
        model="google/gemma-2-9b-it",
      messages=messages,
      temperature=0.6,
      max_tokens=128,
      top_p=0.9
      )
      print(completion.choices[0].message.content)

    except:
      try:

        completion = client.chat.completions.create(
          model="google/gemma-2-9b-it",
        messages=messages,
        temperature=0.6,
        max_tokens=64,
        top_p=0.9
        )
        print(completion.choices[0].message.content)

      except:
        try:

          completion = client.chat.completions.create(
            model="google/gemma-2-9b-it",
          messages=messages,
          temperature=0.6,
          max_tokens=32,
          top_p=0.9
          )
          print(completion.choices[0].message.content)

        except:
          print("Boyut Hatası")

  print("+++---+++")
```
