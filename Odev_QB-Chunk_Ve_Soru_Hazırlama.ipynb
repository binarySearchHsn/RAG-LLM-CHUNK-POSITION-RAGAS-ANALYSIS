```python
filtered_df_2['context_length'] = filtered_df_2['context'].apply(len)
filtered_df_2_smallest_100 = filtered_df_2.nsmallest(100, 'context_length').copy()

filtered_df_5['context_length'] = filtered_df_5['context'].apply(len)
filtered_df_5_smallest_150 = filtered_df_5.nsmallest(150, 'context_length').copy()
```

```python
filtered_df_2_smallest_100 = filtered_df_2_smallest_100.drop(columns=['context_length'])
filtered_df_5_smallest_150 = filtered_df_5_smallest_150.drop(columns=['context_length'])


first_batch = filtered_df_2_smallest_100.iloc[:50].copy()
second_batch = filtered_df_5_smallest_150.iloc[:50].copy()
third_batch = filtered_df_5_smallest_150.iloc[50:100].copy()
fourth_batch = filtered_df_5_smallest_150.iloc[100:150].copy()
fifth_batch = filtered_df_2_smallest_100.iloc[50:100].copy()


result_df = pd.DataFrame(columns=first_batch.columns)

result_df['template'] = pd.NA



#id	question	answer	context	is_negative_response	number_of_articles	ctx_split_points	correct_intro_idx



for i in range(len(first_batch)):

    temp_row=[]

    temp_row.append(i)  #id	
    temp_row.append(third_batch.iloc[i]['question'])
    temp_row.append(third_batch.iloc[i]['answer'])


    temp_split_point=third_batch.iloc[i]['ctx_split_points']
    temp_split_point=ast.literal_eval(temp_split_point)

    temp_split_point.insert(0, 1)

    temp_correct_index= third_batch.iloc[i]['correct_intro_idx']


    temp_context= first_batch.iloc[i]['context'] + second_batch.iloc[i]['context']  
    temp_context= temp_context + third_batch.iloc[i]['context'][temp_split_point[temp_correct_index]:temp_split_point[temp_correct_index+1]]
    temp_context= temp_context+ fourth_batch.iloc[i]['context'] + fifth_batch.iloc[i]['context']

    temp_row.append(temp_context)

    temp_row.append(0)  #is_negative_response

    temp_row.append(15)  #number_of_articles

    temp_ctx_split_points=[]

    temp_ctx_split_points_2 = ast.literal_eval(first_batch.iloc[i]['ctx_split_points']) 


 

    for j in range(len(temp_ctx_split_points_2)):
        if(j!=len(temp_ctx_split_points_2)-1):
            temp_ctx_split_points.append(temp_ctx_split_points_2[j])
        else:
            temp_ctx_split_points.append(len(first_batch.iloc[i]['context']))
    
    end_point= temp_ctx_split_points[-1]

    temp_ctx_split_points_2 = ast.literal_eval(second_batch.iloc[i]['ctx_split_points'])

    for j in range(len(temp_ctx_split_points_2)):
        if(j!=len(temp_ctx_split_points_2)-1):
            temp_ctx_split_points.append(temp_ctx_split_points_2[j]+end_point)
        else:
            temp_ctx_split_points.append(len(second_batch.iloc[i]['context'])+end_point)


    # correct_index_difference = len(temp_split_point[temp_correct_index+1]-temp_split_point[temp_correct_index])
    correct_index_difference = len(third_batch.iloc[i]['context'][temp_split_point[temp_correct_index]:temp_split_point[temp_correct_index+1]])

    end_point= temp_ctx_split_points[-1]
    temp_ctx_split_points.append(correct_index_difference+end_point)
    
    end_point= temp_ctx_split_points[-1]

    temp_ctx_split_points_2 = ast.literal_eval(fourth_batch.iloc[i]['ctx_split_points'])


    for j in range(len(temp_ctx_split_points_2)):
        if(j!=len(temp_ctx_split_points_2)-1):
            temp_ctx_split_points.append(temp_ctx_split_points_2[j]+end_point)
        else:
            temp_ctx_split_points.append(len(fourth_batch.iloc[i]['context'])+end_point)
    

    end_point= temp_ctx_split_points[-1]

    temp_ctx_split_points_2 = ast.literal_eval(fifth_batch.iloc[i]['ctx_split_points'])

    for j in range(len(temp_ctx_split_points_2)):
        if(j!=len(temp_ctx_split_points_2)-1):
            temp_ctx_split_points.append(temp_ctx_split_points_2[j]+end_point)
        else:
            temp_ctx_split_points.append(len(fifth_batch.iloc[i]['context'])+end_point)


    temp_row.append(temp_ctx_split_points)

    temp_row.append(8)  #correct_intro_idx

    template= "Verilen soruyu yalnızca aşağıdaki metne bakarak cevapla. Metin: " + temp_row[3] + ", Soru: " + temp_row[1] + "\n\n\n\n"
    temp_row.append(template)


    result_df.loc[len(result_df)] = temp_row

```

```python
def replace_first_occurrence(original_string, search_string, replace_string):
    first_occurrence_index = original_string.find(search_string)
    if first_occurrence_index == -1:
        return original_string 
    
    new_string = (original_string[:first_occurrence_index] +
                  replace_string +
                  original_string[first_occurrence_index + len(search_string):])
    
    return new_string


def replace_second_occurrence(original_string, search_string, replace_string):
    first_occurrence_index = original_string.find(search_string)
    if first_occurrence_index == -1:
        return original_string  
    
    second_part = original_string[first_occurrence_index + len(search_string):]
    second_occurrence_index = second_part.find(search_string)
    
    if second_occurrence_index == -1:
        return original_string 
    
    second_occurrence_index += first_occurrence_index + len(search_string)
    
    new_string = (original_string[:second_occurrence_index] + 
                  replace_string + 
                  original_string[second_occurrence_index + len(search_string):])
    
    return new_string

```

```python
result_df_first_50=result_df.iloc[0:50].copy()

file_names=['b_icin_1.csv', 'b_icin_2.csv', 'b_icin_3.csv', 'b_icin_4.csv', 'b_icin_5.csv', 'b_icin_6.csv', 'b_icin_7.csv', 'b_icin_8.csv', 'b_icin_9.csv', 'b_icin_10.csv', 'b_icin_11.csv', 'b_icin_12.csv', 'b_icin_13.csv', 'b_icin_14.csv', 'b_icin_15.csv']
files_names_df=['b_icin_df_1.csv', 'b_icin_df_2.csv', 'b_icin_df_3.csv', 'b_icin_df_4.csv', 'b_icin_df_5.csv', 'b_icin_df_6.csv', 'b_icin_df_7.csv', 'b_icin_df_8.csv', 'b_icin_df_9.csv', 'b_icin_df_10.csv', 'b_icin_df_11.csv', 'b_icin_df_12.csv', 'b_icin_df_13.csv', 'b_icin_df_14.csv', 'b_icin_df_15.csv']
for i in range(len(file_names)):
    template=""
    df_df_temp = pd.DataFrame(columns=["template", "correct_answer", "soru"])

    for j in range(len(result_df_first_50)):
        true_split_point_start=result_df_first_50['ctx_split_points'].iloc[j][6]
        true_split_point_end=result_df_first_50['ctx_split_points'].iloc[j][7]

        substring=result_df_first_50["context"].iloc[j][true_split_point_start:true_split_point_end]

        if(i==0):
            new_context_start_index=0
        elif(i<7):
            new_context_start_index=result_df_first_50['ctx_split_points'].iloc[j][i-1]
        else:
            new_context_start_index=result_df_first_50['ctx_split_points'].iloc[j][i]
            
        
        temp_organized_context=result_df_first_50["context"].iloc[j]

        temp_organized_context=temp_organized_context[:new_context_start_index]+substring+temp_organized_context[new_context_start_index:]

        if(i<7):
            organized_context=replace_second_occurrence(temp_organized_context,substring,"")
        else:
            organized_context=replace_first_occurrence(temp_organized_context,substring,"")

        template= template+"Verilen soruyu yalnızca aşağıdaki metne bakarak cevapla. Metin: " + organized_context + ", Soru: " + result_df_first_50["question"].iloc[j] + "\n\n\n\n"


        #yeni_satir = pd.DataFrame({'template': [template], 'correct_answer': [result_df_first_50["answer"].iloc[j]], 'soru': [result_df_first_50["question"].iloc[j]]})
        yeni_satir = pd.DataFrame({'template': [organized_context], 'correct_answer': [result_df_first_50["answer"].iloc[j]], 'soru': ["Verilen soruyu yalnızca metne bakarak cevapla. Soru: "+result_df_first_50["question"].iloc[j]]})
        df_df_temp = pd.concat([df_df_temp, yeni_satir], ignore_index=True)

    df_temptemp = pd.DataFrame([template], columns=["template"])



    df_temptemp.to_csv(file_names[i], index=False)

    df_df_temp.to_csv(files_names_df[i], index=False)
```
